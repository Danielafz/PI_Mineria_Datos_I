Primero importamos las librerías que vamos a usar. pandas para manejar la tabla de datos, numpy para operaciones numéricas, y matplotlib y seaborn para gráficos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Segundo cargamos el archivo CSV en una variable llamada df (de DataFrame, que es como una tabla en pandas) que nos mostrará las primeras 5 filas para ver cómo lucen los datos.

```python
# Carga del dataset
df = pd.read_json('streaming_users_dirty.csv')

# Ver las primeras 5 filas
df.head()
```

## 1. Estructura general del dataset

```python
# Dimensiones del dataset
#En esta celda nos dice cuántas filas y columnas tiene el dataset, y los nombres de cada columna.

print(f"Filas: {df.shape[0]}")
print(f"Columnas: {df.shape[1]}")
print(f"\nNombre de las columnas: {list(df.columns)}")
```

```python
# Tipos de datos de cada columna, por ej. si es un numero entero (int64), número decimal (float64), texto (object) o fecha.
print("Tipos de datos:")
print(df.dtypes)
```

En la celda de arriba observamos que:
* user_id → int64 (número entero)  
* age → int64 (número entero)
* subscription_plan → object (texto)
* monthly_watch_time_mins → float64 (decimal)
* country → object (texto)
* favorite_genre → object (texto)
* last_login_date → object (texto) (en este caso figura como object (texto) cuando debería ser de tipo fecha)
* customer_support_tickets → int64 (número entero)

```python
# Valores faltantes por columna: contamos cuántos valores nulos (vacíos) hay en cada columna.
#Un valor faltante significa que ese campo no tiene dato para ese registro.

print("Valores faltantes por columna:")
print(df.isnull().sum())
print(f"\nTotal de valores faltantes: {df.isnull().sum().sum()}")
```

En este caso encontramos 753 valores faltantes en total distribuidos así:

* monthly_watch_time_mins → 193 nulos
* favorite_genre → 240 nulos
* last_login_date → 320 nulos

```python
# Registros duplicados: Verificamos si hay filas exactamente iguales en el dataset.
#Un duplicado significa que un mismo usuario aparece más de una vez con exactamente los mismos datos.

duplicados = df.duplicated().sum()
print(f"Registros duplicados: {duplicados}")
```

Hay 126 registros duplicados. Eso significa que 126 filas aparecen más de una vez con exactamente los mismos datos.

```python
# Valores únicos por columna categórica
print("Valores únicos en subscription_plan:")
print(df['subscription_plan'].unique())

print("\nValores únicos en country:")
print(df['country'].unique())

print("\nValores únicos en favorite_genre:")
print(df['favorite_genre'].unique())
```

En la celda de arriba nos muestra todos los valores distintos que aparecen en las columnas de texto. Esto nos permite detectar inconsistencias como un mismo valor escrito de distintas formas, por ejemplo "Premium", "PREMIUM", "premium", que deberían ser el mismo.

En el código podemos observar que:
* subscription_plan tiene 15 variantes para solo 3 planes: Estándar, Básico y Premium escritos de formas distintas (mayúsculas, minúsculas, con acento, sin acento, abreviados).
* country tiene los países escritos con nombres completos, abreviados (MEX, ARG, BRA...), en minúsculas, con espacios extra.
* favorite_genre tiene géneros en español e inglés mezclados, en mayúsculas, minúsculas, con errores tipográficos (thriler, accion) y hasta None.

```python
# Estadísticas básicas de columnas numéricas

print("Estadísticas descriptivas de columnas numéricas:")
df[['age', 'monthly_watch_time_mins', 'customer_support_tickets']].describe()
```

En este caso nos muestra el mínimo, máximo, promedio, mediana y otros estadísticos de las columnas numéricas. Nos permite detectar valores imposibles como edades negativas o mayores a 120.
Ademas podemos observar que se confirman varios problemas graves en los datos numéricos:
* age: mínimo -5 y máximo 150 — ambos son imposibles para una edad real.
* monthly_watch_time_mins: mínimo -120 (tiempo negativo, imposible) y máximo 99999 (más de 69 días seguidos viendo contenido, claramente un error).
* customer_support_tickets: mínimo -1 (tickets negativos, imposible) y máximo 150 (valor extremadamente alto).

## Observaciones finales de la inspección inicial

A partir de la inspección que realizamos identificamos los siguientes problemas en el dataset:

- **Dimensiones:** 8.160 filas y 8 columnas.
- **Valores faltantes:** 753 en total. Las columnas afectadas son `monthly_watch_time_mins` (193), `favorite_genre` (240) y `last_login_date` (320).
- **Duplicados:** 126 registros duplicados.
- **Tipo de dato incorrecto:** `last_login_date` figura como texto (object) cuando debería ser de tipo fecha.
- **Inconsistencias en variables categóricas:** `subscription_plan`, `country` y `favorite_genre` presentan múltiples variantes para un mismo valor (mayúsculas, minúsculas, abreviaciones, errores tipográficos).
- **Valores imposibles en variables numéricas:** edades negativas y mayores a 120, tiempos de visualización negativos y extremadamente altos, tickets de soporte negativos.
