```python
import pandas as pd
import numpy as np

# ── CARGA DEL DATASET LIMPIO
df = pd.read_csv('https://raw.githubusercontent.com/Danielafz/PI_Mineria_Datos_I/main/data/processed/streaming_users_clean.csv')
df['last_login_date'] = pd.to_datetime(df['last_login_date'])

print(f"Dataset cargado: {df.shape[0]} filas, {df.shape[1]} columnas")
```

# 1. **Resumen del proceso**

Este proyecto analizó un dataset de usuarios de una plataforma de streaming
que contenía información sobre planes de suscripción, consumo de contenido,
preferencias de género, país de origen y actividad reciente.

*El proceso siguió las siguientes etapas:*

- **Inspección inicial:** se identificaron 8.160 filas, 8 columnas, 753 valores
  faltantes, 126 duplicados, inconsistencias en variables categóricas y valores
  imposibles en variables numéricas.

- **Limpieza y preparación:** se eliminaron duplicados, se normalizaron las
  variables categóricas, se imputaron valores faltantes con moda y mediana
  según el tipo de variable, y se eliminaron registros con valores imposibles.
  El dataset final quedó con 7.003 filas, reteniendo el 85.8% de los datos originales.

- **Análisis exploratorio:** se respondieron 7 preguntas mediante análisis
  univariado, bivariado y multivariado, identificando patrones en el comportamiento
  de los usuarios según su plan de suscripción.

- **Escalamiento y PCA:** se aplicó StandardScaler y PCA sobre 7 variables,
  obteniendo 7 componentes principales con varianza distribuida de forma pareja.

# 2. **Conclusiones por pregunta de análisis**

Las siguientes conclusiones responden a las 7 preguntas definidas en el EDA,
diferenciando la evidencia observada, la interpretación realizada y la conclusión.

* **Pregunta 1 — ¿Cómo se distribuyen los usuarios por plan de suscripción?**

**Evidencia:** el 44.75% de los usuarios tiene plan Básico, el 35.36% Estándar
y el 19.89% Premium.
**Conclusión:** la plataforma tiene una base de usuarios orientada a planes
económicos. Menos de un quinto de los usuarios elige el plan más caro.

* **Pregunta 2 — ¿Cómo es la distribución de edades de los usuarios?**

**Evidencia:** la media de edad es 33.4 años y la mediana es 33 años, con mayor
concentración entre los 25 y 45 años.
**Conclusión:** la plataforma es utilizada principalmente por adultos jóvenes y
de mediana edad, con una distribución aproximadamente simétrica.

* **Pregunta 3 — ¿Qué géneros son los más populares?**

**Evidencia:** Comedia es el género más elegido (17.36%), seguido por Drama
(13.91%) y Acción (13.87%). Documental es el menos elegido (13.62%).
**Conclusión**: los géneros se distribuyen de forma homogénea entre los usuarios.
No existe un género dominante muy marcado, lo que sugiere una audiencia
con gustos variados.

* **Pregunta 4 — ¿Los usuarios con plan Premium ven más contenido?**

**Evidencia:** la mediana de visualización mensual es 1088.88 minutos para
Premium, 838.95 para Estándar y 551.90 para Básico.
**Conclusión:** existe una relación clara entre el plan contratado y el consumo
de contenido. Los usuarios Premium consumen casi el doble que los Básico.

* **Pregunta 5 — ¿Existe relación entre edad y tiempo de visualización?**

**Evidencia:** el coeficiente de correlación entre edad y tiempo de visualización
es 0.003, prácticamente cero.
**Conclusión:** la edad no tiene influencia en el tiempo de visualización. El
consumo de contenido es independiente de la edad del usuario.

* **Pregunta 6 — ¿Cuál es el género favorito según el país?**

**Evidencia**: la distribución de géneros por país es homogénea en todos los
casos, con cada género representando aproximadamente el 14%.
**Conclusión**: no existen preferencias culturales marcadas por país. Los gustos
de los usuarios son similares independientemente de su origen.

* **Pregunta 7 — ¿La edad y el tiempo de visualización varían según el plan?**

**Evidencia**: al analizar edad, tiempo de visualización y plan simultáneamente,
los usuarios Premium se concentran en valores más altos de visualización
independientemente de su edad.
**Conclusión**: el plan de suscripción es el factor más determinante en el consumo
de contenido, por encima de la edad del usuario.

# 3. **Conclusiones del análisis PCA**

**Evidencia:** la varianza se distribuye de forma pareja entre las 7 componentes.
La PC1 explica el 21.83% y las siguientes oscilan entre 13.81% y 14.64%.
Para superar el 80% de varianza acumulada se necesitan 5 componentes y para
el 90% se necesitan 6.

**Interpretación:** la PC1 está dominada por `monthly_watch_time_mins` y
`subscription_plan_enc` (ambas con loading 0.71), confirmando que el perfil
de consumo es la dimensión más relevante del dataset. La PC2 captura el perfil
geográfico y de actividad reciente (`country_enc` y `dias_desde_login`).
La PC3 asocia la edad con las preferencias de género.

**Conclusión:** PCA no logra una reducción dimensional eficiente en este dataset
porque las variables tienen baja correlación entre sí. Sin embargo, confirma
el hallazgo principal del EDA: el plan de suscripción y el tiempo de visualización
son las variables que más información concentran sobre el comportamiento
de los usuarios de la plataforma.

# 4. **Limitaciones y mejoras futuras**

### **Limitaciones**

- El dataset no incluye información sobre ingresos, ocupación ni nivel educativo
  de los usuarios, lo que limita la posibilidad de explicar por qué ciertos
  usuarios eligen planes más costosos.

- El Label Encoding aplicado a variables nominales como `country` y
  `favorite_genre` introduce un orden numérico artificial que no existe en
  los datos reales, lo que puede afectar la interpretación de los componentes
  del PCA.

- La baja correlación entre variables limita la capacidad de PCA para concentrar
  la varianza en pocas componentes, lo que indica que el dataset no tiene una
  estructura multivariada fuerte.

- El alcance de las conclusiones se encuentra condicionado por la información
  disponible y por las decisiones documentadas durante el proceso.

### **Mejoras futuras**

- Una mejora futura podría consistir en incorporar variables adicionales como
  historial de pagos o tiempo de permanencia en la plataforma, que permitirían
  ampliar el alcance del análisis.

- Aplicar One-Hot Encoding en lugar de Label Encoding para las variables
  nominales mejoraría la calidad del análisis PCA.

- Una mejora futura podría consistir en aplicar técnicas de clustering sobre
  los componentes principales para identificar segmentos de usuarios con
  comportamientos similares.
